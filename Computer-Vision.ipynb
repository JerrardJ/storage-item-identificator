```python
# import library
import os
from ultralytics import YOLO
import cv2
```

```python
# Pilih dan load model (sesuaikan sama versi YOLO yang mau dipake)
model = YOLO("yolov8n.yaml")
# model = YOLO("yolov6n.yaml")

# Training
results = model.train(data=os.path.join(ROOT_DIR, "config.yaml"), epochs=250)
```

```python
# inferensi ke video
VIDEOS_DIR = os.path.join(ROOT_DIR, 'videos')

video_path = os.path.join(VIDEOS_DIR, 'testus_instant_noodle.mp4')
video_path_out = os.path.join(VIDEOS_DIR, 'testus_instant_noodle_out_yolov8_146epochs.mp4')  # 5ctmx maksudnya 5 class training model dengan x epoch

cap = cv2.VideoCapture(video_path)
ret, frame = cap.read()
H, W, _ = frame.shape
out = cv2.VideoWriter(video_path_out, cv2.VideoWriter_fourcc(*'MP4V'), int(cap.get(cv2.CAP_PROP_FPS)), (W, H))

model_path = os.path.join(ROOT_DIR, 'runs', 'detect', 'NEW_train_yolov8_all_146epochs', 'weights', 'best.pt')

# Load a model
model = YOLO(model_path)  # load a custom model

threshold = 0.5

while ret:

    results = model(frame)[0]

    for result in results.boxes.data.tolist():
        x1, y1, x2, y2, score, class_id = result

        if score > threshold:
            cv2.rectangle(frame, (int(x1), int(y1)), (int(x2), int(y2)), (0, 255, 0), 4)
            cv2.putText(
                frame,
                results.names[int(class_id)] + " " + str("%.2f" %(results.boxes.conf[0].item() * 100)) + "%",
                (int(x1), int(y1 - 10)), cv2.FONT_HERSHEY_SIMPLEX, 1.3, (0, 255, 0), 3, cv2.LINE_AA
            )

    out.write(frame)
    ret, frame = cap.read()

cap.release()
out.release()
cv2.destroyAllWindows()
```
